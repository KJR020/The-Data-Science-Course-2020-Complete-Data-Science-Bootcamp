```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot  as plt
import seaborn as sns
import sklearn as sk
sns.set()
```

```python
data = pd.read_csv("/content/drive/MyDrive/DataAnalysis_projects/Complete Data Science Training/1.01. Simple linear regression.csv")
data
```

```python
data.describe()
```

```python
y = data["GPA"]
x1 = data["SAT"]
```

```python
plt.scatter(x1,y)
plt.xlabel("SAT",fontsize = 20)
plt.ylabel("GPA",fontsize = 20)
plt.show().
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1+0.275
fig = plt.plot(x1,yhat,lw=4,c="orange",label="regression line")
plt.xlabel("SAT",fontsize=20)
plt.ylabel("GPA",fontsize=20)
plt.show()
```
